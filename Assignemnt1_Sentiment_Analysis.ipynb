```python
import tweepy           # To consume Twitter's API
import pandas as pd     # To handle data
import numpy as np      # For number computing
import re

#function to clean tweets,keep only text
def clean_up_data(raw_text):
    
    # keep only words
    letters_only_text = re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", raw_text)
    #Convert www.* or https?://* to URL
    letters_only_text = re.sub('((www\.[^\s]+)|(https?://[^\s]+))','URL',raw_text)
    #Convert @username to AT_USER
    letters_only_text = re.sub('@[^\s]+','AT_USER',raw_text)
    #Remove additional white spaces
    letters_only_text = re.sub('[\s]+', ' ', raw_text)
    #Replace #word with word
    letters_only_text = re.sub(r'#([^\s]+)', r'\1', raw_text)
    #trim
    letters_only_text = raw_text.strip('\'"')             
    
    # convert to lower case and split 
    words = letters_only_text.lower().split()

    # join the cleaned words in a list
    cleaned_word_list = " ".join(words)

    return cleaned_word_list

def readwords( filename ):
    f = open(filename)
    words = [ line.rstrip() for line in f.readlines()]
    return words

pos =0
neg =0
others =0
total =0
stop =0
positive = readwords('positive.txt')
negative = readwords('negative.txt')
stopwords = readwords('stopwords.txt')

tweets_data_path = 'England.txt'
tweets_data = []
tweets_file = open(tweets_data_path, "r")
for line in tweets_file:
   #read tweet file line by line and clean up the tweets
    tweets_data = clean_up_data(line)
    count = collections.Counter(tweets_data.split())
    #calculate total,positive,negative and stopwords from each tweets
    for key, val in count.most_common():
        total += val
        if key in positive:
            pos += val
        elif key in negative:
            neg += val
        elif key in stopwords:
            stop += val
        else:
            others += val

#count ratio against total word count
pos_ratio = float(pos/total)
neg_ratio = float(neg/total)
others_ratio = float(others/total)
stop_ratio = float(stop/total)


print("Total no. of words:"+ ' ' +str(total))
print("Total no. of positive words:"+ ' '+str(pos))
print("Total no. of negative words:"+ ' '+str(neg))
print("Total no. of other words:"+ ' '+str(others))
print("Total no. of stopwords:"+ ' '+str(stop))
print("The ratio of positive words is " +str(pos_ratio))
print("The ratio of negative words is " +str(neg_ratio))
print("The ratio of other words is " +str(others_ratio))
print("The ratio of stopwords is " +str(stop_ratio))

#check sentiment -whether strongly positive or negative based on threshold 0.3
if pos > neg:
    if pos_ratio>0.03:
        print ("The sentiment is strongly positive for keyword Trump")
    else:
        print("The sentiment is weakly positive for keyword Trump")
    
else:
    if neg_ratio>0.03:
        print ("The sentiment is strongly negative for keyword Trump")
    else:
        print("The sentiment is weakly negative for keyword Trump")


```
